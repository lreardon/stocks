```python
import os
```

```python
directory = '../archive/stocks'

for filename in os.listdir(directory):
		if filename.endswith(".csv") or filename.endswith(".txt"):
			symbol = filename.rsplit('.')[0]
			symbol_lines = open(directory + '/' + filename).readlines()
			nh_symbol_lines = symbol_lines[1:]

			if len(nh_symbol_lines) < 1500:
				continue
			
			if not os.path.exists('ml-datasets/' + symbol):
				os.makedirs('ml-datasets/' + symbol)
			if not os.path.exists('ml-datasets/' + symbol + '/train'):
				os.makedirs('ml-datasets/' + symbol + '/train')
			if not os.path.exists('ml-datasets/' + symbol + '/test'):
				os.makedirs('ml-datasets/' + symbol + '/test')
			if not os.path.exists('ml-datasets/' + symbol + '/validate'):
				os.makedirs('ml-datasets/' + symbol + '/validate')


			training_cutoff = int(len(nh_symbol_lines) * .7)
			testing_cutoff = int(len(nh_symbol_lines) * .85)

			training_lines = [symbol_lines[0]] + nh_symbol_lines[:training_cutoff]
			testing_lines = [symbol_lines[0]] + nh_symbol_lines[training_cutoff:testing_cutoff]
			validating_lines = [symbol_lines[0]] + nh_symbol_lines[testing_cutoff:]

			with open('ml-datasets/' + symbol + '/train/' + symbol + '.csv', 'w') as f:
				f.writelines(training_lines)
			with open('ml-datasets/' + symbol + '/test/' + symbol + '.csv', 'w') as f:
				f.writelines(testing_lines)
			with open('ml-datasets/' + symbol + '/validate/' + symbol + '.csv', 'w') as f:
				f.writelines(validating_lines)
```
